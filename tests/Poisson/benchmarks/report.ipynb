## I. Descriptions

---------------------
### PDE:

\begin{equation}
\frac{\partial u^2}{\partial^2 x} + \frac{\partial u^2}{\partial^2 y} + \frac{\partial u^2}{\partial^2 z} = -12\pi^2\cos{(2\pi x)}\cos{(2\pi y)}\cos{(2\pi z)}
\end{equation}

---------------------
### BCs:

\begin{equation}
\left.\frac{\partial u}{\partial x}\right|_{x=0}=
\left.\frac{\partial u}{\partial x}\right|_{x=1}=
\left.\frac{\partial u}{\partial y}\right|_{y=0}=
\left.\frac{\partial u}{\partial y}\right|_{y=1}=
\left.\frac{\partial u}{\partial z}\right|_{z=0}=
\left.\frac{\partial u}{\partial z}\right|_{z=1}=
0
\end{equation}

---------------------
### Exact solution:

\begin{equation}
u(x, y) = \cos{(2\pi x)}\cos{(2\pi y)}\cos{(2\pi z)}
\end{equation}

## III. Summary

This report shows the performance of AmgX using a 3D Poisson problem for benchmarks run on GW Colonial One clusters. We use one 100M-point and two 4M-point meshes to represent large and small problems, respectively. We used two different 4M-point meshes (Nx=Ny=Nz=160 and Nx=320, Ny=160, Nz=80) in order to understand the effect of mesh geometry, which produces different coefficient-matrix structures. The GPUs we used on the cluster are K20m, and the CPUs are Intel Xeon E5-2670. The network used is InfiniBand.

We conclude that AmgX can provide savings on hardware cost, both from the viewpoint of purchase prize and system management. For example, the run time for solving the 4M-point problem using 4 GPUs is approximately equal to that using 256-CPU cores (16 nodes). One can install 4 GPUs on a single-board workstation or a personal computer, but we need to build a cluster in order to have a 256-CPU-core system. Building a cluster means we need to spend more money on hardware and more time on maintaining and managing the system. Therefore, using AmgX doesn't only provide faster run times but also saves money and time.
